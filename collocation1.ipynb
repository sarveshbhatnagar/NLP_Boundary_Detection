```python
def chi_square(word1, word2):
    
    """
    Tried using two methods, one from book and other from directly calculating. Both gives same result.
    """
    both_holds = bigram_freq[(word1,word2)]
    first_holds = unigrams[word1] - bigram_freq[(word1,word2)]
    second_holds = unigrams[word2]-bigram_freq[(word1,word2)]
    none_holds = sum(bigram_freq.values()) - bigram_freq[(word1,word2)]
    observed = [both_holds, second_holds,first_holds, none_holds]
    
    total = sum(unigrams.values())
    
    pf = expected_prob(word1,total)
    ps = expected_prob(word2,total)
    first = pf*ps*total
    second = (1-pf)*ps*total
    third = pf*(1-ps)*total
    fourth = (1-pf) * (1-ps)*total
    
    expected = [first,second,third,fourth]
    
    chi = 1
    
    for i in range(len(expected)):
        chi += (((observed[i]-expected[i])**2) / expected[i])
            

    return chi
    
    
    
    
```

```python
chi_square("New","York")
```

```python
csli = []
for bg in bigram_freq.keys():
    csli.append((bg,chi_square(bg[0],bg[1])))
```

```python
scli_one = sorted(csli,key=lambda x: x[1],reverse=True)
```

```python
def pmi(word1,word2):
    pw1w2 = (bigram_freq[(word1,word2)]/sum(bigram_freq.values()))
    total = sum(unigrams.values())
    
    pw1 = expected_prob(word1,total)
    pw2 = expected_prob(word2,total)
    
    return math.log(pw1w2/(pw1*pw2))
```

```python
pmi("New","York")
```

```python
pmili = []
for bg in bigram_freq.keys():
    pmili.append((bg,pmi(bg[0],bg[1])))
```

```python
pmili_one = sorted(pmili,key=lambda x: x[1],reverse=True)
```
